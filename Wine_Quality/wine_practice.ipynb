```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
df_red = pd.read_csv('winequality-red.csv', sep=';')
df_white = pd.read_csv('winequality-white.csv', sep=';')
```

```python
print(df_red.shape)
df_red.head()
```

```python
print(df_white.shape)
df_white.head()                          
```

```python
fig, axes = plt.subplots(6, 2, figsize=(12,18), sharey=False)
axes = axes.flatten()
color_palette = sns.color_palette(palette='hls',n_colors=12)
for i , col in enumerate(df_red.columns):
    sns.histplot(df_red[col],kde=True, ax=axes[i], color=color_palette[i])
```
